```python
import pandas as pd
from scipy import stats
from scipy.stats import levene

df = pd.read_csv('drive/MyDrive/multicam_project_data/mbti_all_v2.csv')
df = df.drop('Duration (ms)', axis=1)
```

```python
df['e-i'] = df['mbti'].apply(lambda x: x[0])
df['n-s'] = df['mbti'].apply(lambda x: x[1])
df['t-f'] = df['mbti'].apply(lambda x: x[2])
df['p-j'] = df['mbti'].apply(lambda x: x[3])
```

```python
def vs_mean(group1, group2):
    for column in group1.columns:
        stat, p_value = levene(group1[column], group2[column])
        if p_value < 0.05:
            print(f'등분산 가정이 만족하지 않습니다.')
            # 독립표본 t-검정 수행
            t_stat, p_value = stats.ttest_ind(group1[column], group2[column], equal_var=False)  # equal_var=False는 등분산 가정을 하지 않음
        else:
            print(f'등분산 가정 O')
            # 독립표본 t-검정 수행
            t_stat, p_value = stats.ttest_ind(group1[column], group2[column], equal_var=True)

        if p_value < 0.05:
            # 결과 출력
            print("t-statistic:", t_stat)
            print("p-value:", p_value)
            print(f"{column}에서 통계적으로 유의미한 차이가 있음")
            print('--'*30)
        else:
            print(f"X  {column}에서 통계적으로 유의미한 차이가 없습니다.")
            print('--'*30)
```

# e-i

```python
group1 = df[df['e-i']=='e'].iloc[:,3:15]
group2 = df[df['e-i']=='i'].iloc[:, 3:15]
vs_mean(group1, group2)
```

# n-s

```python
group1 = df[df['n-s']=='n'].iloc[:,3:15]
group2 = df[df['n-s']=='s'].iloc[:, 3:15]
vs_mean(group1, group2)
```

# t-f

```python
group1 = df[df['t-f']=='t'].iloc[:,3:15]
group2 = df[df['t-f']=='f'].iloc[:, 3:15]
vs_mean(group1, group2)
```

## p-j

```python
group1 = df[df['p-j']=='p'].iloc[:,3:15]
group2 = df[df['p-j']=='j'].iloc[:, 3:15]
vs_mean(group1, group2)
```
